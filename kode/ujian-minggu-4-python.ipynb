```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Nama: Prajudi William Chrisdeardo
# Studi Kasus: E-Commerce Sales
# Willi

df = pd.read_csv("D:/myproject/ujian-minggu-4/dataset/ecommerce_sales.csv")
df.isnull().sum()
df
```

```python
df
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')
```

```python
df['Price Each'].fillna(0, inplace=True)
df['Quantity Ordered'].fillna(0, inplace=True)
```

```python
df.info()
```

```python
df['Income']= df['Price Each']*df['Quantity Ordered']
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Price Each'].fillna(0, inplace=True)
df.info()
df['month'] = df['Order Date'].dt.month
```

```python
df['Price Each'].mean()
```

```python
df.info()
```

```python
df['Order ID'].isnull().sum()
df['Order ID'].fillna(0, inplace=True)
df['Order ID'].unique()
```

```python
df['Price Each'].unique()
```

```python
df['Product'].isnull().sum()
df['Product'].fillna('undefined', inplace=True)
df.info()
df['Product'].unique()
```

```python
# Sales paling banyak berada dibulan 12 atau pada akhir tahun
df['Income'].max()
```

```python
df['Income'].min()
```

```python
sns.set(rc={'figure.figsize': (20, 10)})
sns.countplot(x='Income', data=df, hue='month')
plt.xticks(rotation=90)
plt.title('Sales per month')
plt.legend(loc='upper right')
plt.show()
```

```python
# Soal 01
## Pada bulan apakah perusahaan memiliki pendapatan tertinggi dan terendah? Berapa jumlah pendapatan pada bulan tersebut?
## Dari hasil analisa, coba visualisasikan dan berikan pendapat Anda dari visualisasi tersebut.

sns.barplot(x='month', y='Income', data=df)
#xticks untuk rotate huruf di sumbu x nya

plt.xticks(rotation=90)
plt.yticks(rotation = 45)

plt.show
```

```python
data = df['Income'].groupby(df.month).agg(['sum'])
data
plt.plot(data.index, data['sum'], color='red')
plt.grid(ls=':', color='grey')
```

```python
df.info()
```

```python
sales = df['Income'].groupby(df.month).agg('sum').reset_index()
sales.sort_values('Income', ascending=False)
```

```python
# Hasil nomor 1, untuk maksimal berada di bulan ke-12, sedangkan minimal berada di bulan pertama.
sales_max_min = sales[(sales['Income'] == sales['Income'].max()) | ( sales['Income'] == sales['Income'].min())]
sales_max_min
```

```python
sales2 = df['Product'].groupby(df.Income).agg('sum').reset_index()
sales2.sort_values('Income', ascending=False)
```

```python
# Soal 02
## Produk apakah yang laris terjual dan tidak laris terjual?
## Dari hasil analisa, coba visualisasikan dan berikan pendapat Anda dari visualisasi tersebut.

sales_max_min2 = sales2[(sales2['Income'] == sales2['Income'].max()) | ( sales2['Income'] == sales2['Income'].min())]
sales_max_min2
```
